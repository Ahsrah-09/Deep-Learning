```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.info()
```

**Divide dataset into dependent and independent variables**

```python
x=df.iloc[:,3:13]
```

```python
y=df.iloc[:,13]
```

```python
x
```

```python
y
```

**Feature Engineering**

```python
pd.get_dummies(x['Geography'])
```

```python
pd.get_dummies(x['Geography'],drop_first=True)
```

```python
Geography=pd.get_dummies(x['Geography'],drop_first=True)
```

```python
Gender=pd.get_dummies(x['Gender'],drop_first=True)
```

**Concatinate these variable with dataframe**

```python
x=x.drop(['Geography','Gender'],axis=1)
```

```python
x.head()
```

```python
x=pd.concat([x,Geography,Gender],axis=1)
```

```python
x
```

**Splitting the dataset into training set and test set**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import ReLU
from tensorflow.keras.layers import Dropout
```

```python
classifier=Sequential()
```

```python
classifier.add(Dense(units=11,activation='ReLU'))
```

```python
classifier.add(Dense(units=7,activation='ReLU'))
```

```python
classifier.add(Dense(units=6,activation='ReLU'))
```

```python
classifier.add(Dropout(0.2))
```

```python
classifier.add(Dense(1,activation='sigmoid'))
```

```python
classifier.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['Accuracy'])
```

```python
import tensorflow
```

```python
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
model_history=classifier.fit(x_train,y_train,validation_split=0.31,batch_size=10,epochs=800)
```

```python
y_pred=classifier.predict(x_test)
y_pred=(y_pred>=0.5)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
```
